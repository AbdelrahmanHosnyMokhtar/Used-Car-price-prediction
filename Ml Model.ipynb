```python
#trying to removw the outliers
df = df[(df['log_price'] < 13) & (df['log_price'] > 10)]
```

```python
df.columns
```

```python
data_cleaned = df.drop(['Title','Location', 'Posted Date', 'Brand','Ad Description', 'Model', 'Payment Options',
                        'Condition', 'Color','Ad Type', 'Views Count', 'Owner Name', 'Extra Features',
       'Phone Number','Price'],axis = 1)
```

```python
#getting The dummies
data_dummy = pd.get_dummies(data_cleaned,drop_first = True)
data_dummy.shape
```

```python
#Cutting the data into target and input
target = data_dummy['log_price']
inputs = data_dummy.drop('log_price',axis = 1)
```

```python
#scaling the data
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
scaler.fit(inputs)
inputs_scaled = scaler.transform(inputs)
```

```python
#spliting the data
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,  y_test = train_test_split(inputs_scaled,target,test_size = 0.25 , random_state =42)
```

```python
#linear regression model
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_hat = reg.predict(X_train)
```

```python
plt.scatter(y_train,y_hat)
plt.show()
```

```python
sns.distplot(y_train - y_hat);
```

```python
#the score of it
reg.score(X_train,y_train)
```

```python
#The weight of every feature
reg_summry = pd.DataFrame(inputs.columns.values,columns = ['Feature'])
reg_summry['Weights'] = reg.coef_
reg_summry
```

```python
y_hat_test = reg.predict(X_test)
plt.scatter(y_test,y_hat_test,alpha = .2)
plt.show()
```

```python
df_pr = pd.DataFrame(np.exp(y_hat_test),columns=['predictions'])
y_test = y_test.reset_index(drop=True)
df_pr['target'] = np.exp(y_test)
df_pr['diff'] = df_pr['target'] - df_pr['predictions']
df_pr['diff%'] = (df_pr['diff'] / df_pr['target'])*100
df_pr
```

```python
df_pr.describe()
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_hat_test)
```

```python
#random forest model
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 100, random_state = 42)
rf.fit(X_train,y_train)
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(X_test)
# Calculate the absolute errors
errors = abs(predictions - y_test)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2))
```

```python
rf.score(X_train,y_train)
```
